```python
import pandas as pd

EXCEL_FILE = r"C:\Users\Beebrazz\Documents\sales-data-analysis\Superstore.xls"
SHEET_NAME = 0
OUTPUT_SQL = "superstore.sql"
TABLE_NAME = "superstore"
DATABASE_NAME = "sql_sales_analysis"

# Load Excel
df = pd.read_excel(EXCEL_FILE, sheet_name=SHEET_NAME)

# Clean column names (SQL-safe)
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
    .str.replace("-", "_")
)

columns = ", ".join(df.columns)

with open(OUTPUT_SQL, "w", encoding="utf-8") as f:
    f.write(f"USE {DATABASE_NAME};\n\n")

    for _, row in df.iterrows():
        values = []
        for value in row:
            if pd.isna(value):
                values.append("NULL")
            elif isinstance(value, (int, float)):
                values.append(str(value))
            else:
                value = str(value).replace("'", "''")
                values.append(f"'{value}'")

        sql = (
            f"INSERT INTO {TABLE_NAME} ({columns}) "
            f"VALUES ({', '.join(values)});\n"
        )
        f.write(sql)

print("✅ Excel converted to SQL successfully!")
```
